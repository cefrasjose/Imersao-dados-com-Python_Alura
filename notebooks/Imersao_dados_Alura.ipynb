```python
renomear_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
df.rename(columns=renomear_colunas, inplace=True)
df.columns
```

```python
senioridade = {
    'SE': 'senior',
    'MI': 'pleno',
    'EN': 'junior',
    'EX': 'executivo'
}
df['senioridade'] = df['senioridade'].map(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'integral',
    'PT': 'parcial',
    'CT': 'contrato',
    'FL': 'freelancer'
}
df['contrato'] = df['contrato'].map(contrato)
df['contrato'].value_counts()
```

```python
traducao_tamanho_empresa = {
    'M': 'Mdio',
    'L': 'Grande',
    'S': 'Pequeno'
}
df['tamanho_empresa'] = df['tamanho_empresa'].map(traducao_tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
traducao_remoto = {
    0: 'Presencial',
    50: 'Hbrido',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].map(traducao_remoto)
df['remoto'].value_counts()
```

```python
import numpy as np

#Criao de um dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Mariana", "Val"],
    'salario': [4000, np.nan, 5000, np.nan, 10000]
})

#calcula a media salarial e substitui os nulos pela mdia e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#*Calcula Mediana e substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Tera", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 27, 28]
})

df_temperaturas["Preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
import pycountry

# Funo para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

# Criar nova coluna com cdigo ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcularr media salarial por pas (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_paises = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig  = px.choropleth(media_ds_paises,
                     locations='residencia_iso3',
                     color='usd',
                     color_continuous_scale='rdylgn',
                     title='Salrio mdio de Cientista de Dados por Pas',
                     labels={'usd': 'Salrio mdio (USD)', 'residencia_iso3': 'Pas'})
fig.show()
```
